```
# Cell 2: Import libraries and create scraping functions
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_amazon(product):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }

    url = f"https://www.amazon.com/s?k={product.replace(' ', '+')}"

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')
        products = []

        for item in soup.select('.s-result-item'):
            try:
                name = item.select_one('.a-text-normal').get_text(strip=True)
                price_element = item.select_one('.a-price .a-offscreen')

                if price_element:
                    price = float(price_element.get_text().replace('$', '').replace(',', ''))
                    products.append({
                        'name': name[:50] + '...',  # Truncate long names
                        'price': price,
                        'website': 'Amazon'
                    })
            except Exception as e:
                continue
        return products

    except Exception as e:
        print(f"Amazon Error: {str(e)}")
        return []

def scrape_ebay(product):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    url = f"https://www.ebay.com/sch/i.html?_nkw={product.replace(' ', '+')}"

    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        products = []

        for item in soup.select('.s-item__wrapper'):
            try:
                name = item.select_one('.s-item__title').get_text(strip=True).replace('New Listing', '')
                price_element = item.select_one('.s-item__price')

                if price_element:
                    price_text = price_element.get_text().split(' ')[0]
                    price = float(price_text.replace('$', '').replace(',', ''))
                    products.append({
                        'name': name[:50] + '...',
                        'price': price,
                        'website': 'eBay'
                    })
            except Exception as e:
                continue
        return products

    except Exception as e:
        print(f"eBay Error: {str(e)}")
        return []
```

```
# Cell 3: Run the main comparison
def run_comparison():
    product = "wireless headphones"  # You can change this search query
    print(f"Searching for: {product}")

    amazon_results = scrape_amazon(product)
    ebay_results = scrape_ebay(product)

    if not amazon_results and not ebay_results:
        print("No results found. Try again later.")
        return

    all_results = amazon_results + ebay_results
    df = pd.DataFrame(all_results).sort_values(by='price').reset_index(drop=True)

    # Display results with formatting
    print("\nPrice Comparison Results:")
    display(df.head(10))  # Colab-specific display function

    # Save to CSV in Colab's storage
    df.to_csv('price_comparison.csv', index=False)
    print("\nResults saved to price_comparison.csv")

# Execute the comparison
run_comparison()
```
